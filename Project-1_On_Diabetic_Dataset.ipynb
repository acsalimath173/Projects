# Introduction of Dataset
- This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.


# Variables used in the dataset
 - The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.

     - **Pregnancies**               : Number of times pregnant

     - **Glucose**                   : Plasma glucose concentration 2 hours in an oral glucose tolerance test

     - **Blood Pressure**            : Diastolic blood pressure (mm Hg)

     - **Skin Thickness**            : Triceps skin fold thickness (mm)

     - **Insulin**                   : 2-Hour serum insulin (mu U/ml)

     - **BMI**                       : Body mass index (weight in kg/(height in m)^2)

     - **Diabetes Pedigree Function**: Diabetes pedigree function

     - **Age**                       : Age (years)

     - **Outcome**                   : Class variable (0 or 1) 0 mean non-diabetic and 1 means diabetic

## Missing values are filled with KNN method because it should fill with nearest number which is realted to corresponding variables

```python
# Importing sklearn library
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=4)
data_raw_knn = pd.DataFrame(imputer.fit_transform(data_raw))
```

```python
data_raw_knn  # here knn method is imputed for missing values 
```

```python
#Checking missing values before knn and after knn
print("BEFORE_KNN :")
print(data_raw.info())
print("------------------------------------------------------------------------------------------")
print("AFTER_KNN :")
print(data_raw_knn.info())
```

  - Above result we can easily identify the missing values in before and after KNN method

### Finding Outliers in each variables , I used Box-Plot here

```python
#Copying data
dataknn=data_raw_knn.copy()
```

```python
# Finding Outliers in 'Preganancies' variables
import warnings 
warnings.filterwarnings("ignore")
sns.boxplot(dataknn["Pregnancies"])
```

 - Above figure we can see some outliers infront of upper bound so we can say it has rightly skwewed data

```python
# We can find 5 Point Summary for Box-Plot 
# So here I find min,Q1,Q2,Q3,max values
Q1=np.quantile(data_raw["Pregnancies"],0.25)
Q3=np.quantile(data_raw["Pregnancies"],0.75)
IQR=Q3-Q1
Upper_bound = Q3+(1.5*IQR)
Lower_bound = Q1-(1.5*IQR)
print("Q1 s:",Q1)
print("Q3 is:",Q3)
print("IQR is :",IQR)
print("lower_bound is :",Upper_bound )
print("Upper_bound is :",Lower_bound)
Q2=np.quantile(data_raw["Pregnancies"],0.5)
print("Median(Q2) is:",Q2)
```

```python
# Finding Outliers data points
outliers=[]
for feature in data_raw["Pregnancies"]:
    if (feature > Upper_bound) or (feature < Lower_bound):
        outliers.append(feature)
print(outliers)
```
